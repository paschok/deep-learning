## Analysis of the spectrogram

In this tutorial we will familiarise with the spectrogram which is based on the fourier transform. For this purpose we discuss a signal continuous valued one-dimensional signal, e.g., an audio signal. The continuous valued fourier
$$
F(\omega) = \int\limits_{-\infty}^\infty f(t) \exp(i\omega t) \mathrm{d}t
$$
shows how "dominant" a certain frequency $\omega$ is in a given signal $f(t)$, e.g. for a pure sinus signal $f(t) = sin(\omega_0 t)$, $F(\omega)$ becomes inifitely large for $\omega = \omega_0$
$$F(\omega_0) = \int\limits_{-\infty}^\infty \sin(\omega_0t)(\mathrm{i}\sin(\omega_0 t + \cos(\omega_0t))) \mathrm{d}t =\int\limits_{-\infty}^\infty \mathrm{i}\sin^2(\omega_0 t) \mathrm{d}t$$
and zero otherwise.

However, for signal processing we only have a discrete valued signal, sampled from the original signal. The integral above then turns into a sum for the discrete fourier transform (DFT).
$$F(\omega) = \sum\limits_{-n}^\infty f[n] \exp(-\mathrm{i}\omega n t_\mathrm{A})$$ with the sampling frequency $t_\mathrm{A}$.

The trivial implementation of the DFT has a high computational complexity. However the Fast Fourier Transform (FFT) is an algorithm that helps to calculate the DFT quite fast. The algorithm is performed on a subset of the signal, the size of which is relevant for the "resolution" of the frequencies.

A signal using pure tones stored in `pureTones.wav` file. Please, first listen to the signal.

Next, read the signal using the `read` function available in the `scipy.io.wavfile` library. 

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
signal, sr = librosa.load('pureTones.wav')
```

Determine the spectrogram of the signal using the `specgram` function available in the `matplotlib.pyplot` library. 

First, use 256 points to compute the FFT of the signal. For this, adjust the proper variable as argument of the function to compute the spectrogram.

```python
spec = librosa.feature.melspectrogram(y=signal, n_fft = 256, sr=44100)
spec = librosa.power_to_db(spec)
print(spec.shape)
#plt.ylim([0,1000])
plt.imshow(spec)
plt.colorbar()
plt.show()
```

Is the spectrogram you have computed informative? 

Compute the spectrogram again, but now using 4096 points to compute the FFT of the signal.

```python
spec = librosa.feature.melspectrogram(y=signal, n_fft = 4096, sr=44100)
spec = librosa.power_to_db(spec)
#plt.ylim([0,1000])
plt.imshow(spec)
plt.colorbar()
plt.show()
```

What do you think about the spectrogram you have computed now? Can you establish any relationship between the number of points used to compute the FFT and the resolution in the frequency domain of the spectrogram?

The display of the frequencies above is a heat map, i.e., it matches scalar values for the magnitude of a frequence to a pixel with 3 colors. However, it is usually more useful to work with the original scalar values (e.g., as numpy arrays). We recommend to use the librosa library to obtain results.
